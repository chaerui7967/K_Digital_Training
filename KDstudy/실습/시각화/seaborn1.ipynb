```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
snsheadplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.displot(data=df, x="sex", hue="smoker", col="smoker")
# hue = 구분 클래스, 구분 값(범주형으로 넣어야함)
# col = 컬럼
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data = df)
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.groupby('species').count()
```

```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
# 발 길이로 종류를 나눔
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```

```python
sns.scatterplot(x="flipper_length_mm", y = 'body_mass_g',
                hue="species", data = penguins)
```
